```python
import matplotlib.pyplot as plt
import numpy as np
import matplotlib.gridspec as gridspec
import matplotlib.animation as animation

%matplotlib notebook

# sampling variables
start = 100
stop = 1000
step = 100

# data and plot parameter
bins = np.arange(-4, 4, 0.5)
borders = [-4,4,0,1]

# color
axcolor = 'lightgoldenrodyellow'

# generate 4 random variables from the random, gamma, exponential, and uniform distributions
x1 = np.random.normal(0, 1, stop)
x2 = np.random.gamma(1, 1, stop)
x3 = np.random.exponential(1, stop)
x4 = np.random.uniform(-3,3, stop)



# create the function that will do the plotting, where curr is the current frame
def update(c):
    # check if animation is at the last frame, and if so, stop the animation a
    #c = curr*10+100
    #if c == stop: 
    #    a.event_source.stop()
        
    gspec = gridspec.GridSpec(2, 2)

    normal_hist  = plt.subplot(gspec[0, 0])
    gamma_hist   = plt.subplot(gspec[0, 1])
    exp_hist     = plt.subplot(gspec[1, 0])
    uniform_hist = plt.subplot(gspec[1, 1])
    
    
    normal_hist.cla()
    normal_hist.hist(x1[:c], bins=bins, normed=True)
    normal_hist.axis(borders)
    normal_hist.set_title('Normal')
    normal_hist.get_xaxis().set_ticks([])
    
    gamma_hist.cla()
    gamma_hist.hist(x2[:c], bins=bins, normed=True)
    gamma_hist.axis(borders)
    gamma_hist.set_title('Gamma')
    gamma_hist.get_xaxis().set_ticks([])
    
    exp_hist.cla()
    exp_hist.hist(x3[:c], bins=bins, normed=True)
    exp_hist.axis(borders)
    exp_hist.set_title('Exponential')
    
    uniform_hist.cla()
    uniform_hist.hist(x4[:c], bins=bins, normed=True)
    uniform_hist.axis(borders)
    uniform_hist.set_title('Uniform')
    
    plt.annotate('n = {}'.format(c), [-1,0.8])
```

```python
fig = plt.figure()



    
a = animation.FuncAnimation(fig, update, frames=range(start,stop+1,step), interval=200)
#animation.FuncAnimation?
```
