```python
scpop = pd.read_csv('santaclaraco_pop.csv', index_col='Date')
scpop.rename(columns={"Value":"Population"}, inplace=True)
scpop.index = pd.to_datetime(scpop.index)
```

```python
df_zmsp = pd.read_csv('zillow_co13_mspah.csv', index_col='Date')
df_zmsp.rename(columns={"Value":"Median Sold Price"}, inplace=True)
df_zmsp.index = pd.to_datetime(df_zmsp.index)
```

```python
df_sc_imp = pd.read_csv('zillow_co13_IMP.csv', index_col='Date')
df_sc_imp.rename(columns={'Value': 'Housing Inventory'}, inplace=True)
df_sc_imp.index = pd.to_datetime(df_sc_imp.index)
```

```python
# df.resample(rule='A').mean()
df_inventory_year = df_sc_imp.resample(rule='AS').mean()
```

```python
df_mhicsc = pd.read_csv('santa_clara_mhic.csv', index_col='DATE')
df_mhicsc.rename(columns={"VALUE":"Median Household Income"}, inplace=True)
df_mhicsc.index.rename("Date", inplace=True)
df_mhicsc.index = pd.to_datetime(df_mhicsc.index)
```

```python
# df.resample(rule='A').mean()
df_zmsp_year = df_zmsp.resample(rule='AS').mean()
```

```python
santaclaraco = pd.merge(df_zmsp_year, df_mhicsc, how='inner', left_index=True, right_index=True)
```

```python
santaclaraco = pd.merge(santaclaraco, scpop, how='inner', left_index=True, right_index=True)
```

```python
santaclaraco['Population'] = santaclaraco['Population']*1000
```

```python
santaclaraco[['Median Household Income', 'Median Sold Price']].plot(figsize=(8,4), ylim=(0,800000))
```
